```python
from bs4 import BeautifulSoup
```

```python
import lxml
```

```python
with open ("htmls/page1.html","r") as f:
    content = f.read()
```

```python
soup=BeautifulSoup(content,"html.parser")
```

```python
h3=soup.find_all("h3")
```

```python
for h3 in h3:
    print(h3.find("a")["title"])
```

```python
articles=soup.select("article.product_pod")
```

```python
items=[]
for articles in articles :
    title=articles.find("h3").find("a")["title"]
    # print(title)
    price=articles.select_one("p.price_color").text[2:]
    # print(price)
    items.append([title,price])
```

```python
items
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(items,columns=["Books Name","Price in Euro"])
```

```python
df.head()
```

```python
articles=soup.select("article.product_pod")
```

```python
items2=[]
for articles in articles :
    rating_element=articles.select_one("p.star-rating")
    classes=rating_element['class']
    items2.append(classes[1])
    print(classes)
```
