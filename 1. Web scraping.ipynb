```python
for i in range(2,51):
    a=requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    with open(f"htmls/page{i}.html","w",encoding="utf")as f:
        f.write (a.text)
        print(a.status_code)
```
